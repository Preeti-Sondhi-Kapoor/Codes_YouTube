# Set Reset Indexes

```python
df = pd.DataFrame({'A':[1,5,8,5.5],'B':[5,6,8,10],'C':[4,8,'python',9+5j]})
df.set_index('C')        #to make changes permanent, use "inplace=True"
print(df)
```

```python
df['ind']="W X Y Z".split()
df.set_index('ind',inplace=True)        #to make changes permanent, use "inplace=True"
print(df)
```

```python
df.sort_index()
```

```python
df.reset_index()
```

```python
df.reset_index(drop=True)
```
